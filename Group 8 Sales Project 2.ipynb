```python
# Function to check if Promo2 is active for each row
def is_promo2_active(row):
    if row['Promo2'] == 1:
        if (row['Date'].year > row['Promo2SinceYear']) or \
           (row['Date'].year == row['Promo2SinceYear'] and row['Date'].isocalendar().week >= row['Promo2SinceWeek']):
            if row['PromoInterval'] != 'None':
                # Get the month abbreviation (e.g., 'Jan', 'Feb', etc.)
                month = row['Date'].strftime('%b')
                if month in row['PromoInterval'].split(','):
                    return 1
    return 0
```

```python
# Apply the function to the merged_df
merged_df['Promo2Active'] = merged_df.apply(is_promo2_active, axis=1)
```

# EDA

```python
# Load training and store datasets
# 'parse_dates' ensures the Date column is treated as a datetime object
train = pd.read_csv("train.csv", parse_dates=["Date"])
store = pd.read_csv("store.csv")
```

```python
# Merge training data with store data on the Store column
# 'how="left"' keeps all train rows and adds store info where available
df = pd.merge(train, store, on="Store", how="left")
```

```python
# Extract date features for time-based analysis
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Week"] = df["Date"].dt.isocalendar().week
df["Quarter"] = df["Date"].dt.quarter
```

```python
# Create a boolean flag for holidays
# '1' if either StateHoliday is not '0' or it's a SchoolHoliday
df["IsHoliday"] = df["StateHoliday"].astype(str).ne("0") | df["SchoolHoliday"].astype(bool)
```

```python
# Create a flag for whether Promo2 is currently active
# Promo2 is active if the store participates and the PromoInterval is not missing
df["Promo2Active"] = (df["Promo2"] == 1) & df["PromoInterval"].notna()
```

```python
# Calculate the age of each store in months at the time of each sale
# This uses the competition open date and the current sale date
# Missing competition open dates are filled with the current sale date
df["StoreAgeMonths"] = (
    12 * (df["Year"] - df["CompetitionOpenSinceYear"].fillna(df["Year"])) +
    (df["Month"] - df["CompetitionOpenSinceMonth"].fillna(df["Month"]))
).clip(lower=0)
```

```python
# 1. Sales by Quarter
quarterly_sales = df.groupby("Quarter")["Sales"].mean()
quarterly_sales.plot(kind="bar", title="Average Sales by Quarter", ylabel="Average Sales", xlabel="Quarter")
plt.show()
```

## Average Sales by Quarter

This bar chart shows the average sales for each quarter across all stores and years in the dataset. 


Obervations:
- **Quarter 4** tends to generate the **highest average sales**.
- **Quarter 1** shows slightly **lower sales**.

This insight can inform promotional planning during different times of the year.


```python
# 2. Sales by Competition Distance Range
df["CompDistRange"] = pd.cut(df["CompetitionDistance"], bins=[0, 500, 1500, 3000, 6000, np.inf],
                              labels=["<500m", "500-1500m", "1.5-3km", "3-6km", "6km+"])
df.groupby("CompDistRange", observed=False)["Sales"].mean().plot(
    kind="bar", 
    title="Sales by Competition Distance",
    ylabel="Average Sales",
    xlabel="Distance Bin"
)
plt.show()
```

## Sales by Competition Distance

This chart illustrates the relationship between the distance to the nearest competitor and the average sales of a store.

Observations:
- Stores with competitors **less than 500 m** actually show **higher average sales**.
- As competition **distance increases**, average sales **tend to decrease**.

This insight can guide decisions around store placement.


```python
# 3. Boxplot: StoreType vs Sales (with Promo hue)
sns.boxplot(data=df.sample(5000, random_state=42), x="StoreType", y="Sales", hue="Promo")
plt.title("Sales by Store Type & Promo")
plt.show()
```

## Sales by Store Type and Promotion

This boxplot compares the distribution of sales across different store types, segmented by whether a promotion was active (Promo = 1) or not (Promo = 0).

Observations:
- **Store type 'b'** shows the **highest sales**, especially when promotions are active.
- **Store types 'a', 'c', and 'd'** show more **moderate sales** with promotions.
- The interquartile ranges and outliers reveal a large spread in promotional impact.

This visualization supports promoting promotion strategies based on store type.


```python
# 4. Heatmap: Average Sales by Month and Store Type
heatmap_data = df.pivot_table(index="Month", columns="StoreType", values="Sales", aggfunc="mean")
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="YlOrBr")
plt.title("Monthly Average Sales by Store Type")
plt.show()
```

## Monthly Average Sales by Store Type

The heatmap visualizes how average sales vary by month** and store type. Darker shades represent higher sales volumes.

Observations:
- **Store Type 'b'** consistently **outperforms** others in terms of average monthly sales.
- Sales in **Store Type 'a'** and **'d'** follow a relatively **steady** pattern throughout the year.
- The month of **December** shows **peak** sales across all store types.

The performance difference between store types suggests strategies should be considered based on store type.

```python
# 6. Average Sales for stores With vs Without Competition
df["HasCompetition"] = df["CompetitionOpenSinceYear"].notna().astype(int)
avg_sales = df.groupby("HasCompetition")["Sales"].mean()
avg_sales.plot(kind="bar", color=["#1f77b4", "#ff7f0e"])
plt.title("Average Sales: With vs Without Competition")
plt.ylabel("Average Sales")
plt.xticks(ticks=[0, 1], labels=["No", "Yes"], rotation=0)

# Add value labels
for i, val in enumerate(avg_sales):
    plt.text(i, val + 100, f"{val:.0f}", ha='center', fontweight='bold')

plt.show()
```

## Average Sales for Stores With vs Without Nearby Competition

Observations:
- Stores **without competition** appear to have **slightly higher average sales** compared to those with competition.
- The difference may indicate a **modest impact** of nearby competitors.

This comparison is useful for understanding whether the presence of competition is a significant factor in store performance.


```python
# 7. Promo & Promo2 combinations
combo = df.groupby(["Promo", "Promo2"])["Sales"].mean().unstack()
sns.heatmap(combo, annot=True, cmap="Blues", fmt=".0f")
plt.title("Average Sales by Promo and Promo2")
plt.xlabel("Promo2")
plt.ylabel("Promo")
plt.show()
```

## Average Sales by Promo and Promo2

This heatmap visualizes the effect between `Promo` and `Promo2` on average sales.

Observations:
- When **neither Promo nor Promo2** is active (`Promo = 0`, `Promo2 = 0`), the average sales are **4722**.
- When **only Promo2** is active (`Promo = 0`, `Promo2 = 1`), average sale decreases to **4091**.
- When **only Promo** is active (`Promo = 1`, `Promo2 = 0`), average sales increase to **8394**.
- When **both promotions are active** (`Promo = 1`, `Promo2 = 1`), sales are still high at **7588**.

Promo alone may be more effective than Promo2 or the combination.
